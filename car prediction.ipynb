```python
# Importing libraries 
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression


```

```python
#Importing dataset
dataset = pd.read_csv('car_prediction_data.csv')
dataset
```

```python
inputs = dataset.drop(['Car_Name','Owner','Seller_Type',],axis = 'columns')
target = dataset.Selling_Price
target
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
numerics = LabelEncoder()
```

```python
inputs['Fuel_Type_n'] = numerics.fit_transform(inputs['Fuel_Type'])
inputs['Transmission_n'] = numerics.fit_transform(inputs['Transmission'])
inputs
```

```python
#Dropping string columns
inputs_n = inputs.drop(['Fuel_Type','Transmission','Selling_Price'],axis = 'columns')
inputs_n
```

```python
#linear Regression
model = LinearRegression()
```

```python
#Training
model.fit(inputs_n,target)
```

```python
#Prediction
pred = model.predict([[2013,9.54,43000,1,1]])
print(pred)
```
